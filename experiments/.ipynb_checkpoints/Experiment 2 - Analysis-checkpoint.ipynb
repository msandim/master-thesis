## Datasets
11 anomaly detection datasets from the following survey:
- Campos, Guilherme O., et al. "On the evaluation of unsupervised outlier detection: measures, datasets, and an empirical study." Data Mining and Knowledge Discovery 30.4 (2016): 891-927.

## Algorithms:
- LOF
- DBSCAN
- One-class SVM

## Methodology:
- Train/test division (approx. 66% for training and 33% for testing).
- LOF and DBSCAN are fully unsupervised, so they were only applied on the testing data
- One-class SVM was trained using only the inliers of the training data

## How to classify an instance as an outlier?
- LOF assigns a score to each data point, so a threshold needs to be used to make a decision regarding a data instance being outlier or inlier
- DBSCAN assigns all outlier data to a specific cluster (cluster 0 in R's implementation)
- One-class SVM automatically does this in a binary way (this type of model tries to learn the "inlier class" and detect points that should not belong to this class on the test set)

## What parameters were used?
### LOF
- **k parameter**: 3, 5, 8, 14, 19 25, 30
- **threshold**: 1.0 to 5.0 in 0.3 increments (an instance is considered an outlier if the assigned score is higher than this threshold

### DBSCAN
- **Min points**: number of dimensions + 1 (as a suggestion of the documentation in the R package)
- **eps**: 0.3, 0.5, 0.7, 0.9, 1.1

### One-class SVM
Default configurations of the R implementation

## Notes:
In the following data frames, TRUE is assigned as a outlier classification and FALSE otherwise.

```r
setwd("~/master-thesis")
set.seed(20)
library(ggplot2)
library(dplyr)
library(caret)
library(e1071)
library(tidyr)
library(corrplot)
```

```r
results <- readRDS("experiments/results_12Apr_experiment_2.rds")
```

```r
test_lof_thresholds <- function(lof_cols, thresholds)
{
  # Go through each threshold for lof and classify it as outlier or not
  lof_with_thresholds <- lapply(thresholds, function(threshold)
  {
    lof_values <- (lof_cols > threshold) %>% as.data.frame
    names(lof_values) <- paste0(names(lof_values), "_", threshold) # format ex: lof3_1.3
    return(lof_values)
  })
  
  bind_cols(lof_with_thresholds) # join everything in the same dataframe
}

#lof <- data.frame(a = c(5.0, 2.0, 1.1), b = c(2.0, 5.0, 2.3))
#test_lof_thresholds(lof, seq(1.0, 5.0, 0.3))
```

```r
results_processed <- lapply(names(results), function(dname)
{
  dataset <- results[[dname]][[1]]
  
  truth <- (dataset$outlier == "yes")
  
  lof_results <- test_lof_thresholds(dataset %>% select(lof3:lof30), seq(1.0, 5.0, 0.3))
  dbscan_results <- ((dataset %>% select(dbscan03:dbscan11)) == 0) %>% as.data.frame
  svm_results <- (dataset %>% select(svm)) == 1
  
  bind_cols(lof_results, dbscan_results) %>% cbind(svm = svm_results, truth)
})

names(results_processed) <- names(results)
```

# 1 - Analysis on the F1:
- ALOI and KDD are the algorithms with more instances (both inliers and outliers), however those are the datasets in which all techniques have very low F1
- DBSCAN and One-class SVM could not achieve good results (F1 > 0.5) in any of the datasets
- It doesn't seem to exist an optimal LOF configuration that assures a good F1 in every dataset, although this metric is not always available. I examined 2 instances in which F1 had the value of NA/NaN.
- 

```r
f1_results <- sapply(names(results_processed), function(dname)
{
  dataset <- results_processed[[dname]]
  algorithms <- names(dataset)[names(dataset) != "truth"]
  
  sapply(algorithms, function(algorithm_name)
  {
    confusion <- caret::confusionMatrix(dataset[[algorithm_name]] %>% factor(levels = c(TRUE, FALSE)),
                                        dataset$truth %>% factor(levels = c(TRUE, FALSE)),
                                        mode = "prec_recall")
    confusion$byClass[["F1"]]
  })
})
```

```r
f1_results
```

# Making a simple ensemble

```r
results <- readRDS("experiments/results_12Apr_experiment_2.rds")
results_processed <- lapply(names(results), function(dname)
{
  dataset <- results[[dname]][[1]]
  
  truth <- (dataset$outlier == "yes")
  
  lof_results <- dataset %>% select(lof3:lof30)
  dbscan_results <- ((dataset %>% select(dbscan03:dbscan11)) == 0) %>% as.data.frame
  svm_results <- (dataset %>% select(svm)) == 1
  
  bind_cols(lof_results, dbscan_results) %>% cbind(svm = svm_results, truth)
})

names(results_processed) <- names(results)
```

```r
results_processed %>% str
```

```r
library(LogicReg)

results_ensemble <- lapply(names(results_processed), function(dname)
{
  dataset <- results_processed[[dname]]
  
  #dataset$dbscan03 <- factor(dataset$dbscan03, levels = c(TRUE,FALSE))
  #dataset$dbscan05 <- factor(dataset$dbscan05, levels = c(TRUE,FALSE))
  #dataset$dbscan07 <- factor(dataset$dbscan07, levels = c(TRUE,FALSE))
  #dataset$dbscan09 <- factor(dataset$dbscan09, levels = c(TRUE,FALSE))
  #dataset$dbscan11 <- factor(dataset$dbscan11, levels = c(TRUE,FALSE))
  #dataset$svm <- factor(dataset$svm, levels = c(TRUE,FALSE))
  dataset$truth <- factor(dataset$truth, levels = c(TRUE,FALSE))
  
  trainIndex <- createDataPartition(dataset$truth, p = .7, list = FALSE, times = 1)
  
  train <- dataset[trainIndex, ]
  test <- dataset[-trainIndex, ]
  
  print(str(train))
  
  message(paste0("gonna train ", dname))
  model <- train(truth ~ ., data = train, method="glm", family="binomial")
  message("trained")
  
  test$pred <- predict(model, test)
  
  test
})

names(results_ensemble) <- names(results_processed)
```

```r
results_ensemble %>% str
```

```r
f1_results <- sapply(names(results_ensemble), function(dname)
{
  dataset <- results_ensemble[[dname]]
  
  confusion <- caret::confusionMatrix(dataset$pred, dataset$truth, mode = "prec_recall")
  
  message(dname)
  print(confusion)
  
  confusion$byClass[["F1"]]
})
```
