# 1. What results did we achieve on the 11 datasets?

Note: when several parameters were tried (LOF and DBSCAN) we performed voting

```r
results_processed <- lapply(results, function(dataset)
{
  dataset <- dataset %>%
    mutate(vote_lof = (rowMeans(select(dataset, starts_with("lof"))) >= 0.5)) %>%
    mutate(vote_dbscan = (rowMeans(select(dataset, starts_with("dbscan"))) >= 0.5)) %>%
    select(-starts_with("lof"), -starts_with("dbscan"))
})

# Insert the name of each dataset as new column:
results_processed <- lapply(names(results_processed), function(dname)
{
  results_processed[[dname]] <- results_processed[[dname]] %>% mutate(dataset = dname)  
})
```

```r
# Merge everything into a big dataset
results_processed <- do.call("rbind", results_processed) %>% tbl_df
```
